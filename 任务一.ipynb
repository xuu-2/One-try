```python
import numpy as np

# 定义两个 2x2 矩阵
matrix_a = np.array([[1, 2], [3, 4]])
matrix_b = np.array([[5, 6], [7, 8]])

print("=== 矩阵 A ===")
print(matrix_a)
print("\n=== 矩阵 B ===")
print(matrix_b)

# 矩阵加法
print("\n=== 矩阵加法 A + B ===")
print(matrix_a + matrix_b)

# 矩阵乘法
print("\n=== 矩阵乘法 A × B ===")
print(np.dot(matrix_a, matrix_b))

# 矩阵转置
print("\n=== 矩阵 A 的转置 ===")
print(matrix_a.T)
```
